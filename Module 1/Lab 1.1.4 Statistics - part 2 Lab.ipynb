````python
# Calculate the coefficient of best fit line or curve
x = [21,22,23,4,5,6,77,8,9,10,31,32,33,34,35,36,37,18,49,50,100,10,32,49,10,10]
y = [88,12,23,3,6,8,9,12,27,36,34,77,60,35,36,37,18,49,50,100,10,32,49,10,10,22]

print(np.polyfit(x,y,1))
````

````python
# sample points 
X = np.array([0, 5, 10, 15, 20])
Y = np.array([0, 7, 10, 13, 20])
  
# solve for a and b
def best_fit(X, Y):
    
    OutarrXY = np.multiply(X,Y)
    OutarrXX = np.multiply(X,X)
    SumXY = np.sum(OutarrXY)
    SumXX = np.sum(OutarrXX)
    SumX = np.sum(X)
    SumY = np.sum(Y)
    
    n=len(X)
    m=(n*SumXY-(SumX*SumY))/(n*SumXX-(SumX*SumX))
    b=(SumY-(m*SumX))/n
    
    return m, b

# solution
m, b = best_fit(X, Y)
print(f'm = {m} and b = {b}')
print(f'Best fit line y = {m}x + {b}')
plt.plot(X,Y,'o')
plt.plot(X, m*X+b)
plt.show()
````

# 11. [BONUS] Calculate Mean Confidence Interval

Check following methods:

1. scipy.stats.sem
2. scipy.stats.t.ppf

```
# Expected Output:
(30.476190476190474, 19.539561081583383, 41.41281987079756)
```

````python
data = np.array([21,22,23,4,5,6,77,8,9,10,31,32,33,34,35,36,37,18,49,50,100])

def mean_confidence_interval(data, confidence=0.95):
    # Calculate m, h
    m, se = np.mean(data), stats.sem(data)
    h = se * stats.t.ppf((1 + confidence) / 2., len(data)-1)
    return m, m-h, m+h


print(mean_confidence_interval(data))
````
