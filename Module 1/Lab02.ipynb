````python
# Create two arrays of 5 arbitrary numbers 
exampleOne=np.array([25,10,32,15,20])
exampleTwo=np.array([5,15,7,14,8])
barNumber = np.array([2,4,6,8,10])
print(exampleOne)
print(exampleTwo)
print(barNumber)
# Plot two bar charts using plt.bar and the arrays you've created above
# Label them "Example one" and "example two"
fig=plt.figure()
plt.bar(barNumber, exampleOne, label="Example one")
plt.bar(barNumber+1,exampleTwo,label="Example two", color='g')
plt.legend(loc="upper right")
# Label x axis "Bar number"
plt.xlabel("Bar Number")
# Label y axis "Bar height"
plt.ylabel("Bar Height")
# Set title "Bar Graph"
plt.title("Bar Graph")
# Show the chart
plt.show()
````

````python
# Histogram Code 
# Create an array of a random normal distribution
laplaceDistribution = np.random.laplace(loc=15, scale=3, size=100)
# Create a histrogram using plt hist()
n, bins, patches = plt.hist(laplaceDistribution, bins='auto', color='#0504aa',
                            alpha=0.7, rwidth=0.85)
# Set grid for x and y
# set transperancy of the grid using alpha=0.5
plt.grid(axis='both', linestyle='-', linewidth=2, alpha=0.5)
#plt.grid(axis=???, ???)
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('My First Histogram Ever')
plt.text(23, 45, r'$\mu=15, b=3$')
maxfreq = n.max()
# Set a clean upper y-axis limit.
plt.ylim(ymax=np.ceil(maxfreq / 10) * 10 if maxfreq % 10 else maxfreq + 10)
````
