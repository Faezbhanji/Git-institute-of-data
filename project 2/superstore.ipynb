```python
# Mini Project 2
# Import modules needed
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import itertools
import statsmodels.api as stats
# % matplotlib inline
```

```python
# Read the dataset
StoreDataset = pd.read_excel('./Superstore.xls')
StoreDataset.head()
```

```python
StoreDataset.tail()
```

```python
StoreDataset.dtypes
```

```python
# create a list of unique values in the Category variable
set(StoreDataset['Category'])
```

```python
# analyse time series data for category 
furniture_df = StoreDataset.loc[StoreDataset['Category'] == 'Furniture']

# output shows time-series data from 2014 to 2017 for furniture sales.
print(furniture_df['Order Date'].min())
print(furniture_df['Order Date'].max())
furniture_df.head()
```

```python
cols = ['Row ID', 'Order ID', 'Ship Date', 'Ship Mode', 'Customer ID', 'Customer Name', 'Segment', 'Country', 'City','State', 'Postal Code', 'Region', 'Product ID', 'Category', 'Sub-Category', 'Product Name', 'Quantity','Discount', 'Profit']
furniture_df = furniture_df.drop(cols, axis=1)
```

```python
furniture_df.head()
```

```python
furniture_df = furniture_df.groupby('Order Date')['Sales'].sum().reset_index()
```

```python
furniture_df.head()
```

```python
# Now lets set the index to the time series column, the 'order date'
furniture_df = furniture_df.set_index('Order Date')
furniture_df.index
furniture_df.head()
```

```python
# Resample the data so that calculates the monthly value, based on the mean of the daily values for that month
# calculates the mean of the daily values of each month and then assigns that value to the first of each month
MeanMonthlySales = furniture_df['Sales'].resample('MS').mean()
MeanMonthlySales['2017':] #shows the 2017 component of the resampled time series
```

```python
# Lets check to see if this data is stationary by plotting this data to see if there are any patterns
MeanMonthlySales.plot(figsize=(15, 6))
```

```python
# Whilst we can see that the data is clearly seasonal it is not stationary. We can further confirm 
# this by using time series decomposition to visualise trend, seasonality and residual noise, we'll
# use the additive model as the data seems uniformly seasonal
decomposedData = stats.tsa.seasonal_decompose(MeanMonthlySales, model='additive')
decomposedData.plot()
plt.show
```

```python
# above plot shows that setting the frequency of Monthly sales is correct, and the data is not stationary and is 
# obviously Seasonal and 
# the Trend is unstable and fluctuates
```

```python
# now lets have a look at the Seasonal Component of the data as a bar chart
furniture_seasonal = decomposedData.seasonal[0:12] #we only need the first 12 rows of the series
furniture_trend = decomposedData.trend
furniture_trend = furniture_trend.dropna()
# convert the date index values to month
furniture_seasonal.index = furniture_seasonal.index.strftime('%B')
print(furniture_seasonal)
```

```python

furniture_seasonal.plot(kind='bar',title= 'Seasonality Analysis',xlabel='Month', ylabel='Seasonal',figsize=(15, 6),zorder=3)
plt.grid()
```

```python
# Now lets move onto create a Forecasting model. We will use the Sarima model, which is an arima model
# that takes Seasonality into account SARIMA
# Lets generate all the possible parameters for our SARIMA model, this way we can identify the optimal parameters
# to train the model with

p = d = q = range(0, 2) # generate combinations of 0,1
arima_pdq = list(itertools.product(p, d, q))
seasonal_pdq = [ (param_value[0], param_value[1], param_value[2], 12) for param_value in arima_pdq]
print('The parameter combinations for the SARIMA model are:')
for param_arima in arima_pdq:
    for param_seasonal in seasonal_pdq:
        print('{} x {}'.format(param_arima, param_seasonal))
```

```python
for param_arima in arima_pdq:
    for param_seasonal in seasonal_pdq:
        try:
            model = stats.tsa.statespace.SARIMAX(MeanMonthlySales, order=param_arima, seasonal_order=param_seasonal, enforce_stationarity=False,
                                            enforce_invertibility=False)
            results = model.fit()
            print('ARIMA{}x{} - AIC:{}'.format(param_arima, param_seasonal, results.aic))
        except:
            continue
            
```

```python
# the lowest AIC value was AIC:283.3661014370135 -  for SARIMAX (1, 1, 1)x(1, 1, 0, 12). 
# hence we shall use these parameters to train our model.

model = stats.tsa.statespace.SARIMAX(MeanMonthlySales, order=(1,1,1), seasonal_order=(1,1,0,12), 
                                     enforce_invertibility=False)
results = model.fit()
print(results.summary().tables[1])
```

```python
# Let's now run some diagnostics on our model, to check for any unusual patterns
results.plot_diagnostics(figsize=(16,16))
plt.show()
# Whilst not perfect, diagnostics show that the residuals are nearly normally distributed
```

```python
# to see how accurate the results of our model are, we need to validate our forecasts
# compare the actual sales with the predicted sales in the time series data starting from 2016/01/01 
# till the end of the data. We will then check the predicted against what the sales actually were.

forecasted_sales = results.get_prediction(start=pd.to_datetime('2017-01-01'), dynamic=False)
forecasted_sales_conf_int = forecasted_sales.conf_int()
print('This table shows the range of confidence interval for the forecast')
print(forecasted_sales_conf_int)

OverLap=0.7
ax = MeanMonthlySales['2014':].plot(label='observed')
forecasted_sales.predicted_mean.plot(ax=ax, label='One-step ahead Forecast', alpha=OverLap, figsize=(14, 7))
ax.set_xlabel('Date')
ax.set_ylabel('Furniture Sales')

plt.legend()
plt.show()
```

```python
# The above plot is showing the observed values compared to the results of our model. as observed in the plot above 
# the forecasts align with the actual values, showing an upward trend starting at the beginning 
# of the year. It also captured the seasonality pattern nearing the end of the year.
```

```python
# Lets check the Mean Square Error of our predicted values against the actuals
y_predicted = forecasted_sales.predicted_mean
y_actual = MeanMonthlySales['2017-01-01':]
mse = ((y_predicted - y_actual)**2).mean()
print('Mean Square Error is:', round(mse, 2))
```

```python
print('Root Mean Square Error is:', np.sqrt(mse))
```

```python
# The RMSE tells us that our model was able to forecast the average daily furniture sales in the test set 
# within 199 of the real sales. Our furniture daily sales range from around 400 to over 1200. 
# In my opinion, this is a pretty good model so far.
```

```python
# Forecasting the future i.e. out of sample values.
forecast_uc = results.get_forecast(steps=100) # approx 8 years
forecast_ci = forecast_uc.conf_int()
ax = MeanMonthlySales.plot(label='observed', figsize=(14, 6))
forecast_uc.predicted_mean.plot(ax=ax, label='Forecast')
ax.fill_between(forecast_ci.index,
                forecast_ci.iloc[:, 0],
                forecast_ci.iloc[:, 1], color='k', alpha=.25)
ax.set_xlabel('Date')
ax.set_ylabel('Furniture Sales')
plt.legend()
plt.show()
```

```python
# The model has captured the sales seasonality. However as we forecast further into the future, we see that
# the confidence intervals generated by our model, grows.  This is also true of real life, it would be natural 
# to not be too confident of our forecasts
```

```python
furniture_df = StoreDataset.loc[StoreDataset['Category'] == 'Furniture']
office_df = StoreDataset.loc[StoreDataset['Category'] == 'Office Supplies']
furniture_df.shape, office_df.shape
```

```python
#DATA EXPLORATION
 
# We are going to compare the sales for office supplies and furniture in the same time period. 
# By combining two data frames into one and plotting the time series for these two categories into one plot.

cols=['Row ID','Order ID','Ship Date','Ship Mode','Customer ID','Customer Name','Segment','Country','City','State','Postal Code','Region','Product ID','Category','Sub-Category','Product Name','Quantity','Discount','Profit']

furniture_df = furniture_df.drop(cols, axis=1)
office_df = office_df.drop(cols, axis=1)

furniture_df = furniture_df.sort_values('Order Date')
office_df = office_df.sort_values('Order Date')
print(furniture_df.head(5))
print(office_df.head(5))
```

```python
furniture_df = furniture_df.groupby('Order Date')['Sales'].sum().reset_index()
office_df = office_df.groupby('Order Date')['Sales'].sum().reset_index()

furniture_df = furniture_df.set_index('Order Date')
office_df = office_df.set_index('Order Date')

y_furniture = furniture_df['Sales'].resample('MS').mean()
y_office = office_df['Sales'].resample('MS').mean()

furniture_df = pd.DataFrame({'Order Date':y_furniture.index, 'Sales':y_furniture.values})
office_df = pd.DataFrame({'Order Date': y_office.index, 'Sales': y_office.values})
```

```python
# merging the two time series
store_df = furniture_df.merge(office_df, on='Order Date', how='inner')
store_df.rename(columns={'Sales_x': 'Furniture Sales', 'Sales_y': 'Office Sales'}, inplace=True)
```

```python
# chart the office supplies sales and the furniture sales, overlaid

plt.figure(figsize=(15, 8))
plt.plot(store_df['Order Date'], store_df['Furniture Sales'], label='Furniture Sales')
plt.plot(store_df['Order Date'], store_df['Office Sales'], label='Office Supplies')
plt.xlabel('Date')
plt.ylabel('Sales')
plt.title('Furniture & Office Supplies sales')
plt.legend()
```

```python
# the sales of furniture and office supplies share similar seasonal patterns. The sales drop off for both the 
# categories at the begining of the year and picks up once more towards the end of the year. average daily sales is  
# higher for furniture than that of office supplies in most of the months. This is probably, as the value of 
# furniture is higher than those of office supplies. However occasionally, office supplies passed furniture
# average daily sales. Letâ€™s see when was the first time office suppliesâ€™ sales surpassed those of furnitureâ€™s


surpass_date=store_df.loc[store_df['Office Sales']>store_df['Furniture Sales']].head(1)['Order Date']
print('The first time Office Supplies generated more daily sales than furniture: ',surpass_date)       
```

```python
#Lets create a model for forecasting the office supplies
for param_arima in arima_pdq:
    for param_seasonal in seasonal_pdq:
        try:
            model = stats.tsa.statespace.SARIMAX(y_office, order=param_arima, seasonal_order=param_seasonal, enforce_stationarity=False,
                                            enforce_invertibility=False)
            results = model.fit()
            print('ARIMA{}x{} - AIC:{}'.format(param_arima, param_seasonal, results.aic))
        except:
            continue
```

```python
# the lowest AIC value was AIC:305 -  for SARIMAX (1, 1, 1)x(1, 1, 1, 12). 
# hence we shall use these parameters to train our model.

model = stats.tsa.statespace.SARIMAX(y_office, order=(1,1,1), seasonal_order=(1,1,1,12), 
                                     enforce_invertibility=False)
results = model.fit()
print(results.summary().tables[1])
```
